
# Zener Diode Voltage Regulation Analysis

This notebook contains analysis for the Zener diode voltage regulation experiment. It produces two plots:

1. **Voltage Regulation with Variation of Input Voltage (Line Regulation)**
2. **Voltage Regulation with Variation of Load Resistance (Load Regulation)**

The data files should be stored in a `data/` folder in the repository so that the relative paths used below work correctly.



```

import pandas as pd
import matplotlib.pyplot as plt

# Configure plots to be displayed inline
%matplotlib inline
```

## 1) Voltage Regulation with Variation of Input Voltage

We read the input voltage and output voltage data from `data/zener_diode.csv`, then plot $V_{out}$ as a function of $V_{in}$.


```

# Load voltage sweep data
path_voltage = "data/zener_diode.csv"
df_voltage = pd.read_csv(path_voltage)

# Plot V_out vs V_in
plt.figure(figsize=(8, 5))
plt.plot(df_voltage["V_in"], df_voltage["V_out"], marker='o', linestyle='-')
plt.xlabel("Input Voltage (V)")
plt.ylabel("Output Voltage (V)")
plt.title("Voltage Regulation (Variation of Input Voltage)")
plt.grid(True)
plt.show()
```

## 2) Voltage Regulation with Variation of Load Resistance

We read the load resistance and output voltage data from `data/zener_diode_resistance.csv`, then plot $V_{out}$ as a function of the load resistance $R_{\mathrm{L}}$.


```

# Load load-resistance sweep data
path_resistance = "data/zener_diode_resistance.csv"
df_resistance = pd.read_csv(path_resistance)

# Plot V_out vs Load Resistance
plt.figure(figsize=(8, 5))
plt.plot(df_resistance["load_resistance"], df_resistance["output_voltage"], marker='s', linestyle='-')
plt.xlabel("Load Resistance (Ω)")
plt.ylabel("Output Voltage (V)")
plt.title("Voltage Regulation (Variation of Load Resistance)")
plt.grid(True)
plt.show()
```
